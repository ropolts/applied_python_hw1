```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

# 1. Анализ исторических данных

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('temperature_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])

def analyze_city_temperature(df_city, print_mode=False):
    city_name = df_city['city'].iloc[0]
    
    df_city = df_city.sort_values('timestamp').reset_index(drop=True)
    df_city['rolling_mean'] = df_city['temperature'].rolling(window=30, min_periods=1, center=True).mean()
    df_city['rolling_std'] = df_city['temperature'].rolling(window=30, min_periods=1, center=True).std()
    df_city['is_anomaly'] = abs(df_city['temperature'] - df_city['rolling_mean']) > 2 * df_city['rolling_std']
    
    seasonal_temp_profile = df_city.groupby('season')['temperature'].agg(average='mean', std='std').reset_index()
    
    trend_regr = LinearRegression().fit((df_city.index + 1).values.reshape(-1, 1), df_city['temperature'])
    trend_type = 'warming' if trend_regr.coef_[0] > 0 else 'stability' if  trend_regr.coef_[0] == 0 else 'cooling'
    df_city['trend'] = trend_regr.predict((df_city.index + 1).values.reshape(-1, 1))
    
    min_city_temperature = df_city['temperature'].min()
    max_city_temperature = df_city['temperature'].max()
    avg_city_temperature = df_city['temperature'].mean()
    
    if print_mode:
        print('--- General information ---')
        print(f'City: {city_name}')
        print(f'Minimum temperature: {min_city_temperature:.2f}°C')
        print(f'Average temperature: {avg_city_temperature:.2f}°C')
        print(f'Maximum temperature: {max_city_temperature:.2f}°C')
        print(f'Trend: {trend_type} (avg {'+' if trend_regr.coef_[0] > 0 else '' if trend_regr.coef_[0] == 0 else '-'}{trend_regr.coef_[0] * 365:.2f}°C/year)\n')
        
        print('--- Seasonal temperature profile ---')
        print(seasonal_temp_profile.to_string(index=False))
        print('\n')
        
        plt.figure(figsize=(12, 6))
        plt.plot(df_city['timestamp'], df_city['temperature'], label='Temperature')
        upper_bound = df_city['rolling_mean'] + 2 * df_city['rolling_std']
        lower_bound = df_city['rolling_mean'] - 2 * df_city['rolling_std']
        plt.fill_between(df_city['timestamp'], lower_bound, upper_bound, color='lightblue', label=r'2$\sigma$ CI')
        plt.plot(df_city['timestamp'], df_city['trend'], label='Trend', color='orange', linestyle='--')
        anomalies = df_city[df_city['is_anomaly']]
        plt.scatter(anomalies['timestamp'], anomalies['temperature'], color='red', label='Anomalies')
        plt.title(f'Temperature in {city_name}')
        plt.xlabel('Date')
        plt.ylabel('Temperature (°C)')
        plt.grid()
        plt.legend()
        plt.tight_layout();
        return None
    return {'city_name': city_name,
           'min_city_temperature': min_city_temperature,
           'avg_city_temperature': avg_city_temperature,
           'max_city_temperature': max_city_temperature,
            'trend_type': trend_type,
            'seasonal_temp_profile': seasonal_temp_profile,
            'anomalies': df_city[df_city['is_anomaly']]
           }
```

# 2. Мониторинг текущей температуры

```python
import requests
from datetime import datetime

api_key = '22731cda6059b9ac77c4fbd5a7ac0cb3'

def get_current_temperature(city, api_key=api_key):
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'
    response = requests.get(url)
    data = response.json()
    try:
        season = month_to_season[datetime.fromtimestamp(data['dt']).month]
        temp = data['main']['temp']
        return temp, season
    except:
        try:
            msg = data['message']
            return msg
        except:
            return 'Error occured'

def check_anomaly(city, cur_temp, season, df=df, print_mode=False):
    df_city = df[df['city'] == city]
    seasonal_temp_profile = analyze_city_temperature(df_city)['seasonal_temp_profile'].set_index('season')
    avg_season_temp, std_season_temp = seasonal_temp_profile.loc[season]
    is_anomaly = abs(cur_temp - avg_season_temp) > 2 * std_season_temp
    if print_mode:
        print('--- Current weather report ---')
        print(f'City: {city}')
        print(f'Current temperature: {cur_temp}°C')
        print(f'Average season temperature: {avg_season_temp:.2f}°C (± {2 * std_season_temp:.2f}°C)')
        print(f'Is anomaly: {is_anomaly}\n')
        return None
    return is_anomaly
```

```python
city_list = df.city.unique()
for city in city_list:
    temp, season = get_current_temperature(city)
    check_anomaly(city, temp, season, print_mode=True)
```

### Комментарий
- В большинстве городов температура находится в пределах средней сезонной температуры.
- Аномалия обнаружена только в Сиднее.

### Сравнение синхронного и асинхронного подхода

### Синхронный подход

```python
%%time
for city in city_list:
    temp, season = get_current_temperature(city)
    check_anomaly(city, temp, season)
```

### Асинхронный подход

```python
import aiohttp
import asyncio
import nest_asyncio

async def get_current_temperature_async(city, session, api_key=api_key):
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'
    async with session.get(url) as response:
        data = await response.json()
        try:
            season = month_to_season[datetime.fromtimestamp(data['dt']).month]
            temp = data['main']['temp']
            return temp, season
        except:
            try:
                msg = data['message']
                return msg
            except:
                return 'Error occured'

async def main():
    nest_asyncio.apply()
    async with aiohttp.ClientSession() as session:
        tasks = [get_current_temperature_async(city, session) for city in city_list]
        results = await asyncio.gather(*tasks)
        for city, result in zip(city_list, results):
            temp, season = result
            check_anomaly(city, temp, season)
```

```python
%%time
asyncio.run(main())
```

### Комментарий
Выполнение большого числа сетевых запросов является I/O-bound задачей. Асинхронный подход позволил обрабатывать запросы к API одновременно, что привело к значительному росту скорости выполнения.
